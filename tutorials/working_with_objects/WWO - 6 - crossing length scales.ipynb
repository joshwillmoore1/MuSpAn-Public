<h1>Combining points into a shape - in progress</h1>

One of the core aims of MuSpAn is to enable truly <i>multiscale</i> spatial analysis. One means of doing that is to provide spatial analysis tools which operate across length scales, such as methods from Topological Data Analysis or certain Spatial Statistics. However, there is another way of crossing scales in MuSpAn - converting between shape-like and point-like data.
